```python
import nsepy
import datetime
import pandas as pd
```

```python
def gethdata(stockname):
    today=datetime.date.today()
    duration=300
    duration2=44
    start=today+datetime.timedelta(-duration)
    end1=today
    stockdata=nsepy.get_history(symbol=stockname,start=start,end=end1)
    return stockdata
rf=gethdata("ITC")
```

```python
rf.tail()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics 
```

```python
ndh=rf[['High']]
v=ndh.values
v=np.append(v, [0])
v=np.delete(v,0)
v
```

```python
ndh2=rf[['Low']]
v2=ndh2.values
v2=np.append(v2, [0])
v2=np.delete(v2,0)
v2
```

```python
rf['ndhigh']=v
rf['ndlow']=v2
```

```python
rf.reset_index(inplace = True)
```

```python
rf.isnull().sum()
```

```python
X= rf.drop(['Date','ndhigh','Symbol','Series','ndlow'],axis=1)
Y=rf['ndhigh']
X2= rf.drop(['Date','ndhigh','Symbol','Series','ndlow'],axis=1)
Y2=rf['ndlow']
```

```python
Y.head()
```

```python
Xtestn=X.iloc[-1:]
Xtestn
```

```python
Xtestnl=X2.iloc[-1:]
Xtestnl
```

```python
X.drop(index=X.index[-1], 
        axis=0, 
        inplace=True)
Y.drop(index=Y.index[-1], 
        axis=0, 
        inplace=True)
```

```python
X2.drop(index=X2.index[-1], 
        axis=0, 
        inplace=True)
Y2.drop(index=Y2.index[-1], 
        axis=0, 
        inplace=True)
```

```python
len(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=3)
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, Y2, test_size = 0.1, random_state=3)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor2 = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

```python
regressor2.fit(X2_train,Y2_train)
```

```python
test_data_prediction = regressor.predict(X_test)
```

```python
test_data_prediction2 = regressor2.predict(X2_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared error : ", error_score)
```

```python
Y_test = list(Y_test)
```

```python
Y2_test = list(Y2_test)
```

```python
plt.subplots(figsize=(20, 10))
plt.plot(Y_test,'-o', color='blue', label = 'Actual high Value')
plt.plot(test_data_prediction,'-o', color='red', label='Predicted high Value')
plt.plot(Y2_test,'-x', color='green', label = 'Actual low Value')
plt.plot(test_data_prediction2,'-x', color='orange', label='Predicted low Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('HIGH & LOW Price')
plt.legend()
plt.show()
```
